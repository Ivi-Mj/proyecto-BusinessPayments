```python
df1 = pd.read_csv(os.path.join(dataset_path, 'extract - cash request - data analyst.csv'))
df2 = pd.read_csv(os.path.join(dataset_path, 'extract - fees - data analyst - .csv'))
```

# **Inner join**

```python
df_unido = pd.merge(df1, df2, on='id', how='inner')
print(df_unido)
```

# **Left join**

```python
df_unido = pd.merge(df1, df2, on='id', how='left')
print(df_unido)
```

# **Right Join**

```python
df_unido = pd.merge(df1, df2, on='id', how='right')
print(df_unido)
```

# **Outter join**

```python
df_unido = pd.merge(df1, df2, on='id', how='outer')
print(df_unido)
```
